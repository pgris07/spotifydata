```python
df_tracks["release_date"] = pd.to_datetime(df_tracks["release_date"],format='ISO8601')
df_tracks.set_index("release_date", inplace=True)
df_tracks.head()
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks
```

```python
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
corr_df = df_tracks.select_dtypes(include=['float64', 'int64']).drop(["key", "mode", "explicit"], axis=1)
corr_df = corr_df.corr(method="pearson")


plt.figure(figsize=(15, 8))
plt.title("Corelation HeatMap between variables")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

sns.heatmap(corr_df, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y = "loudness", x = "energy", line_kws={"color":"red"}, color = "firebrick")
plt.title("Loudness VS Energy Co-relation");
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y = "popularity", x = "acousticness",line_kws={"color":"red"}, color = "royalblue")
plt.title("Loudness VS Energy Co-relation");
```

```python
df_tracks ["dates"] = df_tracks.index.get_level_values ("release_date")
df_tracks["dates"] = pd.to_datetime(df_tracks["dates"])
years = df_tracks["dates"].dt.year
```

```python
df_tracks.head()
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist", kde=True, color='blue')
plt.title("Songs released per year");
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.barplot(x= years, y = df_tracks["duration"], ax=ax, errorbar= None, color= "royalblue")
plt.xticks(rotation=90, fontsize =12)
plt.title("Years vs Duration");
```
